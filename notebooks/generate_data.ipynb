```python
import random
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
inputs = 5
outputs = 2
time = 5000
time_list = np.arange(0,time)
n_reps = 5
a_r = 0.3

out = np.zeros((inputs+outputs,time))
cols = []
for i in range(inputs):
    w = 2.0*np.pi*3.0*random.random()*n_reps/time
    ph = -np.pi + 2.0*np.pi*random.random()
    A = -1.0 + 2.0*random.random()
    x = np.sin(w*time_list + ph)
    out[i] = A*(np.sin(w*time_list + ph) + a_r*np.random.rand(time))
    cols.append('in_' + str(i))
    for o in range(outputs):
        A = -1.0 + 2.0*random.random()
        out[-1-o] = out[-1-o] + A*out[i]   

for o in range(outputs):
    cols.append('out_' + str(o))
```

```python
plt.plot(out[-1])
plt.plot(out[-2])
plt.show()
```

```python
out_df = pd.DataFrame(out.T,columns=cols)
out_df.to_csv("../data/1.raw/regression.csv",index=False)
```
